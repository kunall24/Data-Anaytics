## 1. What is a random variable in probability theory?

* A random variable is basically a way to assign numbers to outcomes of a random experiment. So instead of just saying "heads or tails", we can say 0 for heads and 1 for tails. It helps in analyzing things mathematically.

## 2. What are the types of random variables?

* There are mainly two types
  1. Discrete random variable.
  2. Continuous random variable

  Discrete random variables take specific values (like 1, 2, 3…), while continuous ones can take any value in a range (like 1.23, 2.75, etc.).

## 3. What is the difference between discrete and continuous distributions?

* Discrete distributions deal with variables that have separate values, like number of students in a class. Continuous distributions are used when the variable can be any value within a range, like height, weight or temparature.

## 4. What are probability distribution functions (PDF)?

* PDFs describe how likely it is for a random variable to take a certain value or fall within a range. For continuous variables, PDF is more about probability density, not actual probabilities for exact values.

## 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

* While PDF tells us how dense the probability is at a point, the CDF gives the total probability up to a certain point. So, it adds up all the probability up to that value.

## 6. What is a discrete uniform distribution?

* In a discrete uniform distribution, all possible values have equal chance of happening. Like rolling a fair die, every number from 1 to 6 has the same probability (1/6).

## 7. What are the key properties of a Bernoulli distribution?

* A Bernoulli distribution only has two outcomes: success (1) and failure (0). It’s the simplest case of probability, like flipping a coin once.

## 8. What is the binomial distribution, and how is it used in probability?

* The binomial distribution is used when we do a certain experiment multiple times (like flipping a coin 10 times), and we want to find the probability of getting a specific number of successes. It’s basically repeated Bernoulli trials.

## 9. What is the Poisson distribution and where is it applied?

* Poisson distribution is used when we want to model rare events that happen over a time or space, like number of emails in an hour or accidents on a road per month. It’s good when events happen randomly but at a known average rate.

## 10. What is a continuous uniform distribution?

* A continuous uniform distribution is when every value in a range has equal chance of happening. Like picking a random number between 0 and 1 any number in that range is equally likely.

## 11. What are the characteristics of a normal distribution?

* A normal distribution looks like a bell curve it’s symmetric and most of the values are near the mean. The mean, median, and mode are all the same in this distribution. It’s one of the most common ones in stats.

## 12. What is the standard normal distribution, and why is it important?

* The standard normal distribution is a special case of the normal distribution where the mean is 0 and standard deviation is 1. It’s useful because it makes calculations easier and is used in many statistical tests.

## 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

* CLT says that if you take a large enough sample from any population, the average of those samples will follow a normal distribution, no matter what the original distribution was. This is super useful in stats because it lets us use normal distribution even if the original data isn’t normal.

## 14. How does the Central Limit Theorem relate to the normal distribution?

* It’s related because CLT basically explains why the normal distribution shows up so often. Thanks to CLT, the averages of sample data become normally distributed as the sample size grows, which helps in many statistical methods.

## 15. What is the application of Z statistics in hypothesis testing?

* Z-statistics help us check how far our sample mean is from the population mean. In hypothesis testing, it helps to see if the difference is big enough to be considered significant or just happened by chance.

## 16. How do you calculate a Z-score, and what does it represent?

* Z-score is calculated by taking the difference between a value and the mean, and dividing it by the standard deviation. It tells us how many standard deviations away a value is from the mean.

## 17. What are point estimates and interval estimates in statistics?

* A point estimate is just a single value guess for a population parameter, like the sample mean. An interval estimate gives a range where we think the actual value lies this is more reliable than just a single number.

## 18. What is the significance of confidence intervals in statistical analysis?

* Confidence intervals show the range where we expect the true population value to lie. They also give us a confidence level (like 95%), which means we’re pretty sure the true value is in that range.

## 19. What is the relationship between a Z-score and a confidence interval?

* Z-scores are used to build confidence intervals. For example, to make a 95% confidence interval, we use the Z-score that captures the middle 95% of the normal distribution.

## 20. How are Z-scores used to compare different distributions?

* Z-scores help compare values from different distributions by standardizing them. So even if two datasets have different scales or means, we can compare how extreme a value is using Z-scores.

## 21. What are the assumptions for applying the Central Limit Theorem?

* CLT assumes that the samples are independent, identically distributed, and that the sample size is large enough. Usually, 30 or more is considered good for the CLT to apply.

## 22. What is the concept of expected value in a probability distribution?

* Expected value is like the average you’d get if you repeat an experiment many times. It’s the sum of all possible values multiplied by their probabilities. It gives an idea of the long-term result.

## 23. How does a probability distribution relate to the expected outcome of random variables?

* The probability distribution tells us all the possible outcomes and their chances. When we use that info to calculate the expected value, we get a sense of what result we can expect over time or many repetitions.
