#**Theoritical Questions and Answers**


## 1. What is hypothesis testing in statistics?
   
  * Hypothesis testing is a way to check if there's enough evidence in a sample of data to support a particular claim about a population. It helps us make decisions using data instead of just guessing.

## 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

  * The null hypothesis (H₀) usually says that there's no effect or no difference. The alternative hypothesis (H₁) is what you believe might be true instead — that there is an effect or a difference. Hypothesis testing checks if the data gives us enough reason to reject the null.

## 3. What is the significance level in hypothesis testing, and why is it important?
  * The significance level (α), often set at 0.05, is the probability of rejecting the null hypothesis when it's actually true (Type I error). It tells us how strict we want to be in deciding what counts as enough evidence.

## 4. What does a P-value represent in hypothesis testing?
  * The p-value shows the probability of getting results at least as extreme as what we observed, assuming the null hypothesis is true. A small p-value means the observed result is unlikely under H₀.

## 5. How do you interpret the P-value in hypothesis testing?
   
  * If the p-value is less than the significance level (say, 0.05), we reject the null hypothesis — it means the results are statistically significant. If it’s more, we don’t have enough evidence to reject H₀.

## 6. What are Type 1 and Type 2 errors in hypothesis testing?
   
  * A Type 1 error happens when we wrongly reject a true null hypothesis. A Type 2 error is when we fail to reject a false null hypothesis. Basically, Type 1 = false alarm, Type 2 = missed detection.

## 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
   
  * A one-tailed test checks for a difference in only one direction (greater than or less than), while a two-tailed test checks for any difference (either greater or smaller). Use two-tailed when you don’t know the direction beforehand.

## 8. What is the Z-test, and when is it used in hypothesis testing?
   
  * A Z-test is used when the population standard deviation is known, and the sample size is large (usually n > 30). It’s useful for comparing sample and population means.

## 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
   
  * The Z-score = (X - μ) / σ. It tells us how many standard deviations a value is from the mean. In testing, it helps find probabilities and p-values using the standard normal distribution.

## 10. What is the T-distribution, and when should it be used instead of the normal distribution?
    
  * The T-distribution is used when the sample size is small (n < 30) or the population standard deviation is unknown. It's similar to the normal curve but has heavier tails to account for more variability.

## 11. What is the difference between a Z-test and a T-test?
    
  * The Z-test uses population standard deviation and large samples; the T-test uses sample standard deviation and small samples. T-tests are more common when dealing with real-world data.

## 12. What is the T-test, and how is it used in hypothesis testing?
    
  * A T-test compares means to see if they are significantly different. It’s used in situations like checking if two groups (like control and test) have different average outcomes.

## 13. What is the relationship between Z-test and T-test in hypothesis testing?
    
  * Both are used to test means, but the Z-test is more precise when we know population parameters. When those are unknown (which is often), we go with the T-test, especially for small samples.

## 14. What is a confidence interval, and how is it used to interpret statistical results?
    
  * A confidence interval gives a range within which the true population parameter is likely to fall. If a 95% confidence interval for the difference in means doesn’t include 0, it means there’s likely a significant difference.

## 15. What is the margin of error, and how does it affect the confidence interval?
    
  * The margin of error tells how much we expect our estimate to vary. A bigger margin means a wider confidence interval — more uncertainty. A smaller one means more precision.

## 16. How is Bayes' Theorem used in statistics, and what is its significance?
    
  * Bayes' Theorem updates the probability of an event based on new evidence. It’s powerful in real-world decision making, like in medical testing or spam filters, where you need to adjust beliefs based on results.

## 17. What is the Chi-square distribution, and when is it used?
    
  * The Chi-square distribution is used in tests involving categorical data, like checking if observed counts match expected ones. It's skewed right and depends on degrees of freedom.

## 18. What is the Chi-square goodness of fit test, and how is it applied?
    
  * It checks if a sample’s distribution fits a known distribution. You compare observed vs. expected frequencies using the formula: χ² = Σ((O-E)²/E). If χ² is large, there’s likely a mismatch.

## 19. What is the F-distribution, and when is it used in hypothesis testing?
    
  * The F-distribution is used when comparing two variances or in ANOVA tests. It’s always positive and skewed, and helps check if group variances are significantly different.

## 20. What is an ANOVA test, and what are its assumptions?
    
  * ANOVA (Analysis of Variance) checks if there are significant differences between more than two group means. Assumptions: normal distribution, equal variances, and independent samples.

## 21. What are the different types of ANOVA tests?
    
  * There’s one-way ANOVA (one factor), two-way ANOVA (two factors), and repeated measures ANOVA (same subjects tested multiple times). Each has different use cases depending on study design.

## 22. What is the F-test, and how does it relate to hypothesis testing?
  
  * An F-test is used to compare variances or test group differences in ANOVA. It tells if the variation between groups is more than what we’d expect by chance.

## 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = [102, 98, 100, 101, 99, 103, 97, 100]
sample_mean = np.mean(sample)
population_mean = 100
population_std = 2
n = len(sample)

# Z-test
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))  # two-tailed

print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

## 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2(mu0=0, mu1=1, sigma=1, alpha=0.05):
    x = np.linspace(-4, 6, 1000)
    y0 = norm.pdf(x, mu0, sigma)
    y1 = norm.pdf(x, mu1, sigma)

    # Critical value
    z_crit = norm.ppf(1 - alpha)

    plt.plot(x, y0, label='H₀: μ=0', color='blue')
    plt.plot(x, y1, label='H₁: μ=1', color='orange')

    plt.fill_between(x, y0, where=(x > z_crit), color='red', alpha=0.4, label='Type I Error (α)')
    plt.fill_between(x, y1, where=(x <= z_crit), color='green', alpha=0.4, label='Type II Error (β)')

    plt.axvline(z_crit, linestyle='--', color='black')
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.xlabel("Test Statistic")
    plt.ylabel("Density")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

plot_type1_type2()
```

## 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = np.std(data, ddof=1) / np.sqrt(n)
    margin = t.ppf((1 + confidence) / 2, df=n-1) * stderr
    return (mean - margin, mean + margin)

data = np.random.normal(120, 15, 30)
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci[0]:.2f} to {ci[1]:.2f}")
```

## 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
# Prior = P(A), Likelihood = P(B|A), Evidence = P(B)
def bayes_theorem(prior, likelihood, evidence):
    return (likelihood * prior) / evidence

# Example:
# P(Disease) = 0.01, P(Pos | Disease) = 0.95, P(Pos) = 0.05
posterior = bayes_theorem(0.01, 0.95, 0.05)
print(f"Posterior Probability: {posterior:.4f}")
```

## 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

# Observed and expected frequencies
observed = [50, 30, 20]
expected = [40, 40, 20]

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")
```
