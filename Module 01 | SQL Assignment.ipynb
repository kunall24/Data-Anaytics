## 5. Explain the consequences of attempting to insert, update, or delete data in a way that violates constraints. Provide an example of an error message that might occur when violating a constraint.

### _Consequences_ :

* INSERT: If you insert data violating a constraint (e.g., duplicate in a primary key), the row will not be added.

* UPDATE: If you try to update a value that breaks a constraint (e.g., making a foreign key point to a non-existing parent record), the update fails.

* DELETE: If you delete a row that is referenced by a foreign key, the deletion will be blocked (unless ON DELETE CASCADE is set).

### _Example error messages_:

* Primary Key violation:

  ERROR: duplicate key value violates unique constraint "students_pkey"
  DETAIL: Key (student_id)=(101) already exists.


* Foreign Key violation:

  ERROR: insert or update on table "orders" violates foreign key constraint "orders_customerid_fkey"
  DETAIL: Key (customer_id)=(123) is not present in table "customers".


* Check constraint violation:

  ERROR: new row for relation "employees" violates check constraint "salary_check"
  DETAIL: Failing row contains (emp_id=5, name='Raj', salary=-5000).

## 1. Identify the primary keys and foreign keys in mavenmovies db. Discuss the differences.


#### Differeences between Primary and Foreign Key:

  <u>Primary Key:</u>
  * Primary key column/attribute of a table/an entity takes values which are unique and are not null.

  <u>Foreign Key:</u>
  * Foreign Key column/attribute of a table/an entity are use dto create link between two tables. Where, the table which is using another table as reference is calles as Child table and the table which is being Referenced is called Parent table.
  * The link is only possible when a primary colum is used from paratent table for referencing whereas child table can have duplicate values in referenced (foreign) column.
  * The foreign key column in child table cannot have a value which is not part of Parent tables referenced column. The query will throw error.
